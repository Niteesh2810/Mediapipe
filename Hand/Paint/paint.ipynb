```python
import cv2
import mediapipe as mp
import import_ipynb
import hand_landmarks_module1 as hlm
import numpy as np
import math
```

```python
hand=hlm.handDetector()
```

```python
cap=cv2.VideoCapture(0)

drawColor=(0,255,0)
option=None
ix,iy=0,0
canvas=np.zeros((480,640,3),np.uint8)
pen_thickness=5
eraser_thickness=5
flag=None


while True:
    ret,frame=cap.read()
    img=cv2.flip(frame,1)
    hand.findHand(img)
    lmList=hand.findPosition(img,draw=False)
    
    if len(lmList)!=0:
        fing,label=hand.fingersUpAdv(img,lmList)
        
        if label=='Right':           
            
            if len(fing)!=0 and fing[0]==0 and fing[1]==1 and fing[2]==1 and fing[3:]==[0,0]:
                #print('Select')
                option='Select'
                cv2.putText(img,"Select ",(50,50),cv2.FONT_HERSHEY_COMPLEX,1,(255,255,0),2)
              
            
            elif len(fing)!=0 and fing[0]==0 and fing[1]==1 and fing[2:]==[0,0,0]:
                #print('Move')
                option='Move'
                
                if option=='Move' and label=='Right':
                    x,y=lmList[8][1],lmList[8][2]
                    if ix==0 and iy==0:
                        ix,iy=x,y
                        
                    ix,iy=x,y
                    
                    
                    if drawColor==(0,0,0):
                        cv2.line(img,(ix,iy),(x,y),drawColor,eraser_thickness)
                        cv2.line(canvas,(ix,iy),(x,y),drawColor,eraser_thickness)
                    
                    else:
                        cv2.line(img,(ix,iy),(x,y),drawColor,pen_thickness)
                        cv2.line(canvas,(ix,iy),(x,y),drawColor,pen_thickness)
                            
            
            elif len(fing)!=0 and fing[0]==0 and fing[1:5]==[1,1,1,1]:
                #print('Adjust')
                option='Adjust'
                flag=0
                cv2.putText(img,"Adjust Eraser Thickness",(50,50),cv2.FONT_HERSHEY_COMPLEX,1,(255,255,0),2) 
            
            elif len(fing)!=0 and fing[0]==1 and fing[1]==1 and fing[2]==1 and fing[3:]==[0,0]:
                option='Adjust'
                flag=1
                cv2.putText(img,"Adjust Pen Thickness",(50,50),cv2.FONT_HERSHEY_COMPLEX,1,(255,255,0),2)
        
        
        
        elif option=='Select' and label=='Left':
            if len(fing)!=0 and fing[0]==0 and fing[1]==1 and fing[2:]==[0,0,0]:
                drawColor=(255,0,0)
            
            elif len(fing)!=0 and fing[0]==0 and fing[1]==1 and fing[2]==1 and fing[3:]==[0,0]:
                drawColor=(0,255,0)
            
            elif len(fing)!=0 and fing[0]==1 and fing[1]==1 and fing[2]==1 and fing[3:]==[0,0]:
                drawColor=(0,0,255)
                
            elif len(fing)!=0 and fing[0]==0 and fing[1:5]==[1,1,1,1]:
                drawColor=(0,0,0) 
                
                
        elif option=='Adjust' and label=="Left":
            x1,y1=lmList[4][1],lmList[4][2]
            x2,y2=lmList[8][1],lmList[8][2]
            dist=math.sqrt(((x2-x1)**2)+((y2-y1)**2))
            cv2.line(img,(x1,y1),(x2,y2),(0,255,0),2)
            
            if flag==1:
                pen_thickness=int(np.interp(dist,(20,170),(1,20)))
                #print(dist,pen_thickness)
                cv2.putText(img,str(pen_thickness),(50,200),cv2.FONT_HERSHEY_COMPLEX,1,(0,255,0),3)
                
            else:
                eraser_thickness=int(np.interp(dist,(20,170),(1,20)))
                #print(dist,eraser_thickness)
                cv2.putText(img,str(eraser_thickness),(50,200),cv2.FONT_HERSHEY_COMPLEX,1,(0,255,0),3)
            
        
        else:
            ix,iy=0,0
    
           
    if option=='Select':
        cv2.putText(img,"Select",(50,50),cv2.FONT_HERSHEY_COMPLEX,1,(255,255,0),2)   
    
    elif option=='Adjust':
        
        if flag==1:
            cv2.putText(img,"Adjust Pen Thickness",(50,50),cv2.FONT_HERSHEY_COMPLEX,1,(255,255,0),2)
            cv2.putText(img,str(pen_thickness),(50,200),cv2.FONT_HERSHEY_COMPLEX,1,(0,255,0),3)
        else:
            cv2.putText(img,"Adjust Eraser Thickness",(50,50),cv2.FONT_HERSHEY_COMPLEX,1,(255,255,0),2)
            cv2.putText(img,str(eraser_thickness),(50,200),cv2.FONT_HERSHEY_COMPLEX,1,(0,255,0),3)
    
    
    # Code for getting the drawing on the screen
    imgGray=cv2.cvtColor(canvas,cv2.COLOR_BGR2GRAY)
    _,imgInv=cv2.threshold(imgGray,50,255,cv2.THRESH_BINARY_INV)
    imgInv=cv2.cvtColor(imgInv,cv2.COLOR_GRAY2BGR)
    img=cv2.bitwise_and(img,imgInv)
    img=cv2.bitwise_or(img,canvas)
    
    cv2.imshow('PAINT',img)
    #cv2.imshow('canvas',canvas)
    if cv2.waitKey(1) & 0xFF==ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

### RIGHT HAND:

    Index                       ----> Move
    
    Index + Middle              ----> Select
    
    Thumb + Index + Middle      ----> Adjusting pen thickness
    
    Thumb alone closed (No 4)   ----> Adjusting eraser thickness
    
    
    


### LEFT HAND


<b><u>Colors:

Index                       ----> Blue

Index + Middle              ----> Green

Thumb + Index + Middle      -----> Red

Thumb alone closed (No 4)   ----> Eraser


<b><u>Thickness:

Use Thumb and Index finger to adjust thickness
